### **Load Data**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

data = '/content/drive/MyDrive/Mechine Learning/wbc.csv' # path dataset
df = pd.read_csv(data) # load dataset
```

## 1.  Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

Kolom Unnamed: 32 tidak memiliki pengaruh pada data karena seluruh nilainya adalah null atau kosong. Dan kolom id tidak relevan untuk analisis karena hanya merupakan pengenal unik untuk setiap entri dan tidak mengandung informasi yang dapat digunakan untuk klasifikasi kanker.

```python
df = df.drop(columns=['Unnamed: 32'])
df = df.drop(columns=['id'])  # Menghapus kolom id karena tidak diperlukan
```

```python
df.head()  # Cek hasil setelah slicing
```

## Cek kolom yang belum numerik

```python
# Memeriksa kolom numerik
def check_numeric_columns(df):
    non_numeric_columns = [col for col in df.columns if not pd.api.types.is_numeric_dtype(df[col])] #jika ada kolom yang belum numerik akan dimasukkan kedalam non_numeric_columns
    return non_numeric_columns

# Memeriksa kolom
non_numeric_columns = check_numeric_columns(df)

# Menampilkan hasil
if non_numeric_columns:
    print("Ada kolom yang belum numerik:", non_numeric_columns)
else:
    print("Semua kolom sudah numerik.")
```

## 2. Lakukan proses encoding pada kolom "diagnosis".

Hal ini karena model pembelajaran mesin membutuhkan data dalam bentuk numerik.

```python
# mengganti nilai M (Malignant) = 1 dan B (Benign) = 0

le = LabelEncoder()  # Membuat objek dari LabelEncoder
df['diagnosis'] = le.fit_transform(df['diagnosis'])  # Proses encoding diagnosis
df.head()  # Cek hasil setelah encoding
```

##3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

melakukan standarisasi pada semua kolom numerik kecuali kolom target diagnosis. Standarisasi bertujuan agar semua fitur berada pada skala yang sama dengan mean 0 dan standar deviasi 1.

```python
# Membuat objek dari StandardScaler
# Standarisasi berarti mengubah data sehingga memiliki rata-rata 0 dan deviasi standar 1.
# Ini membantu dalam memastikan bahwa fitur-fitur memiliki skala yang sama dan mempercepat konvergensi algoritma machine learning.

std = StandardScaler()
numeric_columns = df.columns.drop('diagnosis')  # Kolom numerik kecuali diagnosis karena kolom ini adalah kolom target, yang dimana jika diubah akan menganggu proses
df[numeric_columns] = std.fit_transform(df[numeric_columns])  # Standarisasi fitur
df.head()  # Verifikasi hasil standarisasi
```

## 4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

Untuk melatih model, kita perlu membagi dataset menjadi data latih (80%) dan data uji (20%). Kita akan menggunakan train_test_split dengan stratified split agar distribusi kelas diagnosis tetap seimbang.

```python
df_train, df_test = train_test_split(
    df,                       # Dataset asli yang akan dibagi.
    test_size=0.2,            # 20% dari data asli akan menjadi data uji.
    random_state=0,           # memastikan bahwa pembagian data adalah konsisten setiap kali kode dijalankan dengan nilai ini.
    stratify=df['diagnosis']  # Stratifikasi untuk distribusi kelas yang seimbang
)
print(f'B = 0')
print(f'M = 1')
print('==========================')
print(f'Jumlah data asli: {df.shape[0]}')
print(f'Jumlah data train: {df_train.shape[0]}')
print(f'Jumlah data test: {df_test.shape[0]}')

print('==========================')
print(f'Jumlah label data asli:\n{df["diagnosis"].value_counts()}')
print(f'Jumlah label data train:\n{df_train["diagnosis"].value_counts()}')
print(f'Jumlah label data test:\n{df_test["diagnosis"].value_counts()}')
```
